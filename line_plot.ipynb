## Line Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
x = [2,3,4,5,6,7,8]
y = [3,5,7,6,7,8,9]
plt.plot(x,y)
```

```python
plt.plot(x, y, color='r' )
```

```python
plt.plot(x, y, color='r',marker = 'o' )
```

```python
plt.plot(x, y, color='r',marker = 'o', linestyle = '--' )
```

```python
plt.plot(x, y, color='r',marker = 'o', linestyle = '--',linewidth = 3 )
```

```python
plt.plot(x, y, color='r',marker = 'o', linestyle = '--',linewidth = 3, markersize = 10 )
```

```python
plt.figure(figsize=(12,9))
plt.plot(x, y, color='r',marker = 'o', linestyle = '--',linewidth = 3, markersize = 10, label='Line-Plot' )
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.title('Line Plot')
plt.legend()
```

```python
z=[3,4,6,5,6,7,8]
plt.figure(figsize=(12,9))
plt.plot(x, y, color='r',marker = 'o', linestyle = '--',linewidth = 3, markersize = 10, label='Line-Plot1' )
plt.plot(x, z, color='m',marker = '*', linestyle = ':',linewidth = 3, markersize = 10, label='Line-Plot2' )
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.title('Line Plot')
plt.legend()
```
